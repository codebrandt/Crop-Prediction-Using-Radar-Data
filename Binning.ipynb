```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
from matplotlib import pyplot
import numpy as np
from scipy.stats import binned_statistic
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import KBinsDiscretizer
```

```python
data_path = "/content/drive/MyDrive/ADM/WinnipegDataset_PCA_GA_revised.csv"
df_data = pd.read_csv(data_path)
df_data.head()
```

```python
d_data = df_data.copy()
d_data.head()
```

```python
del df_data['label']
df_data
```

```python
cols = df_data.columns
cols
```

```python
trans = KBinsDiscretizer(n_bins=3, encode='ordinal', strategy='uniform')
# trans = KBinsDiscretizer(n_bins=3, encode='ordinal', strategy='quantile')
data = trans.fit_transform(df_data)
```

```python
f = open("/content/drive/MyDrive/ADM/bin_range_PCA_GA_revised_3.csv", "w")
f.write("0,1,2,3\n")
for i in range(18):
    s = ",".join(list(map(str, trans.bin_edges_[i]))) + "\n";
#     print(s)
#     break
    f.write(s)
f.close()
```

```python
dataset = pd.DataFrame(data, columns = cols)
dataset
```

```python
dataset['label'] = d_data['label'].values
dataset
```

```python
cols = dataset.columns.tolist()
cols
```

```python
cols = cols[-1:] + cols[:-1]
cols
```

```python
ddataset = dataset[cols]
ddataset
```

```python
ddataset.to_csv('/content/drive/MyDrive/ADM/WinnipegDataset_PCA_GA_revised_3.csv', index=False)
```
