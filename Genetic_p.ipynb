```python
from sklearn.metrics import *
from sklearn.model_selection import *
from sklearn.preprocessing import * 
from sklearn.linear_model import LogisticRegression
import pandas as pd
!pip install sklearn-genetic
import numpy as np
from genetic_selection import GeneticSelectionCV
```

```python
mcc = make_scorer(matthews_corrcoef)
estimator = LogisticRegression(solver = "liblinear", C = 6, tol = 1, fit_intercept = True)
```

```python
report = pd.DataFrame()
nofeats = [] 
chosen_feats = [] 
cvscore = [] 
rkf = RepeatedStratifiedKFold(n_repeats = 2, n_splits = 10)
```

```python
data_path = "/content/drive/MyDrive/ADM/data_ga_r.csv"
# pd_data = pd.read_csv(data_path)
D = pd.read_csv(data_path)
# pd_data.shape
```

```python
# D = pd_data.iloc[:len(pd_data) // 2,:]
D.shape
```

```python
allfeats = list(D.columns)
allfeats.remove('label')
numcols = [i for i in allfeats]
```

```python
D[allfeats].dtypes
```

```python
for col in numcols: 
  scaler = MinMaxScaler()
  D[col] = scaler.fit_transform(D[col].values.reshape(-1,1))
```

```python
D.iloc[:,-1] = D.iloc[:,-1].astype(np.int64)
D[allfeats]
```

```python
X = D[allfeats] ; y = D["label"].values
y
```

```python
for i in range(2, 102):
  
  selector = GeneticSelectionCV(estimator,
                                cv = rkf,
                                verbose = 0,
                                scoring = mcc,
                                max_features = i,
                                n_population = 50,
                                crossover_proba = 0.5,
                                mutation_proba = 0.1,
                                n_generations = 10,
                                crossover_independent_proba=0.5,
                                mutation_independent_proba=0.05,
                                #tournament_size = 3,
                                n_gen_no_change=10,
                                caching=True,
                                n_jobs=-1)
  selector = selector.fit(D[allfeats], y)
  genfeats = D[allfeats].columns[selector.support_]
  genfeats = list(genfeats)
  print("Chosen Feats:  ", genfeats)

  cv_score = selector.generation_scores_[-1]
  nofeats.append(len(genfeats)) 
  chosen_feats.append(genfeats) 
  cvscore.append(cv_score)
  print(i)
report["No of Feats"] = nofeats
report["Chosen Feats"] = chosen_feats
report["Scores"] = cvscore
```

```python
Chosen_Feats_op =  ['f3', 'f4', 'f8', 'f11', 'f26', 'f28', 'f34', 'f50', 'f58', 'f83', 'f93', 'f101', 'f103', 'f122', 'f140', 'f143', 'f155', 'f167']
```
